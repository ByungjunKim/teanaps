```python
from Teanaps.TextAnalytics.TfidfCalculator import TfidfCalculator

PATH = "sample2.txt"
POS_LIST = ["NNG", "NNP"]

tokenized_sentence_list = []
f = open(PATH, encoding="utf-8")
for line in f:
    line = line.strip()
    words = [w.split("/")[0] for w in line.split(" ") if w.split("/")[3] in POS_LIST]
    new_line = ""
    for word in words:
        new_line += " " + word
    tokenized_sentence_list.append(new_line.strip())
    f.flush()
f.close()

tfidf = TfidfCalculator()
tfidf.calculation_tfidf(tokenized_sentence_list)
```
