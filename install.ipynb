```python
from teanaps.nlp.SyntaxAnalyzer import SyntaxAnalyzer

sa = SyntaxAnalyzer()

document_list = []
tokenized_sentence_list = []

POS_LIST = ["NNG", "NNP"]

f = open("data/article_sample.txt", encoding="utf-8")

for line in f:
    line = line.strip()
    col = line.split("\t")
    label = col[0]
    source = col[1]
    datetime = col[2]
    title = col[3]
    content = col[4]
    document_list.append([label, source, datetime, title, content])
    tagged_word_list = sa.parse(content)
    tokenized_sentence = ""
    for word, pos in tagged_word_list:
        if pos in POS_LIST:
            tokenized_sentence += " " + word
    tokenized_sentence_list.append(tokenized_sentence.strip())
f.close()

print(tagged_word_list)

tokenized_sentence_list[:5]
```
